### DATA INFO
- Administrative:사용자 계정 관리 관련 활동을 위해 사용자가 방문한 페이지 수
- Administrative Duration: 사용자가 관리 페이지에서 보낸 시간
- Informational:웹사이트에 대해 사용자가 방문한 페이지 수
- Informational Duration:사용자가 정보 페이지에 머문 시간
- Product Related:사용자가 방문한 제품 관련 페이지 수
- Product Related Duration:사용자가 제품 관련 페이지에서 보낸 시간
- Bounce Rates: 사용자가 방문한 페이지의 평균 이탈률. 해당 페이지를 통해 사이트에 들어왔다가 같은 사이트 내의 다른 페이지를 계속 보지 않고 떠나는 방문자의 비율.
- Exit Rates:사용자가 방문한 페이지의 평균 이탈률. 해당 페이지에서 사이트를 떠난 사람들의 비율.
- Page Values:사용자가 방문한 평균 페이지 수. 사용자가 목표 페이지에 도달하거나 전자상거래(또는 둘 다)를 완료하기 전에 방문한 페이지의 평균 값.
- Special Day: 어머니의 날과 같은 특별한 행사나 크리스마스와 같은 축제에 가까운 방문일.
- Month: 방문한 달.
- Operating Systems: 방문자의 운영 체제.
- Browser:방문자의 브라우저.
- Region:방문자가 세션을 시작한 지리적 지역입니다.
- Traffic Type: 사용자가 웹사이트에 들어온 트래픽 소스스
- Visitor Type: 사용자의 재구매여부
- Weekend: 주말이용 여부
- Revenue 수익을 냈는지

```python
drop = df.query('Administrative == 0 and Informational == 0 and ProductRelated == 0').index
df.drop(drop, inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# duration 계산해서 채우기기
adm_mean = df.Administrative.mean()
adm_du_mean = df.Administrative_Duration.mean()
adm = adm_mean / adm_du_mean
df.loc[(df.Administrative != 0) & (df.Administrative_Duration == 0), 'Administrative_Duration'] = df.loc[(df.Administrative != 0) & (df.Administrative_Duration == 0), 'Administrative'] * adm

inf_mean = df.Informational.mean()
inf_du_mean = df.Informational_Duration.mean()
inf = inf_mean / inf_du_mean
df.loc[(df.Informational != 0) & (df.Informational_Duration == 0), 'Informational_Duration'] = df.loc[(df.Informational != 0) & (df.Informational_Duration == 0), 'Informational'] * inf

pro_mean = df.ProductRelated.mean()
pro_du_mean = df.ProductRelated_Duration.mean()
pro = pro_mean / pro_du_mean
df.loc[(df.ProductRelated != 0) & (df.ProductRelated_Duration == 0), 'ProductRelated_Duration'] = df.loc[(df.ProductRelated != 0) & (df.ProductRelated_Duration == 0), 'ProductRelated'] * pro
```

```python
# 타겟과 피처 데이터 관계 시각화화
# 각 카테고리별로 subplot에 그릴 stacked bar 그래프 생성
categories = ['Region', 'TrafficType', 'Weekend', 'Month', 'SpecialDay','VisitorType']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 8))

for i, category in enumerate(categories):
    row, col = i // 3, i % 3
    ax = axes[row][col]

    # 해당 카테고리별로 true/false count 계산
    revenue_count = df.groupby([category, 'Revenue']).size().unstack(fill_value=0)

    # stacked bar plot 그리기
    revenue_count.plot(kind='bar', stacked=True, ax=ax)

    # x, y 축 레이블, 그래프 타이틀 설정
    ax.set_xlabel(category)
    ax.set_ylabel('Count')
    ax.set_title(f'Revenue by {category}')

plt.tight_layout()
plt.show()
```

```python
#비율로 살펴보기
# subplot 초기화
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# 컬럼 리스트
columns = ['Region', 'TrafficType', 'Weekend', 'Month', 'SpecialDay','VisitorType']

# for loop로 subplot에 그래프 그리기
for col, ax in zip(columns, axes.flatten()):
    # 전체를 100%로 두고 true/false 비율 계산
    df_true = df[df['Revenue'] == True].groupby(col).count() / df.groupby(col).count()
    df_false = df[df['Revenue'] == False].groupby(col).count() / df.groupby(col).count()

    # 막대 차트 그리기
    ax.bar(df_true.index, df_false['Revenue'], label='False')
    ax.bar(df_true.index, df_true['Revenue'], bottom=df_false['Revenue'], label='True')
    ax.legend()
    ax.set_title(f'Revenue Ratio by {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Revenue Ratio')

# subplot 간격 조정
plt.tight_layout()
plt.show()
```

```python
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'June': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

df['Month'] = df['Month'].map(month_mapping)
```

```python
# 불필요한 column 삭제하기
df_cleaned=df.drop(columns=['Administrative','Informational','ProductRelated','OperatingSystems','Browser','BounceRates'])
df_cleaned.reset_index(drop=True,inplace=True)
df_cleaned.shape
```

## 데이터구분

```python
y = df_cleaned['Revenue']
print('타겟데이터의 비율\n', y.value_counts(normalize=True))

%matplotlib inline
sns.countplot(x=y);
```

```python
# baseline score
baseline_score = df_cleaned['Revenue'].value_counts(normalize=True).max()
print('Baseline Score',baseline_score)
```

```python
# 데이터 분리
target = 'Revenue'
y = df_cleaned[target]
X = df_cleaned.drop(target,axis=1)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
X_train,X_val,y_train,y_val = train_test_split(X_train,y_train,test_size=0.2,stratify=y_train,random_state=42)
```

```python
print(X_train.shape,X_val.shape,X_test.shape)
```

## 데이터 학습

```python
# 로지스틱 회귀
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *

# scaling
from sklearn.preprocessing import RobustScaler
numeric_feats = X_train.dtypes[X_train.dtypes != 'object'].index

scaler = RobustScaler()
X_train[numeric_feats] = scaler.fit_transform(X_train[numeric_feats])
X_val[numeric_feats] = scaler.transform(X_val[numeric_feats])

# encoding
from category_encoders import OneHotEncoder
ohe = OneHotEncoder()

X_train_ohe = ohe.fit_transform(X_train)
X_val_ohe = ohe.transform(X_val)

# 학습
logistic = LogisticRegression(class_weight={0:0.15,1:0.85})
logistic.fit(X_train_ohe,y_train)
```

```python
print('logistic훈련', logistic.score(X_train_ohe,y_train))
print('logistic검증', logistic.score(X_val_ohe, y_val),'\n')
lg_pred = logistic.predict(X_val_ohe)
print(confusion_matrix(y_val,lg_pred))
print(classification_report(y_val,lg_pred))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier(random_state=42,class_weight='balanced')
dt.fit(X_train_ohe, y_train)

dt_pred = dt.predict(X_val_ohe)
print("dt훈련: ", dt.score(X_train_ohe,y_train))
print("dt검증: ", dt.score(X_val_ohe,y_val), '\n')
print(confusion_matrix(y_val,dt_pred))
print(classification_report(y_val, dt_pred))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42, 
    n_jobs=-1,
    class_weight='balanced')
rf.fit(X_train_ohe, y_train)

print("rf훈련 정확도: ", rf.score(X_train_ohe, y_train))
print("rf검증 정확도: ", rf.score(X_val_ohe, y_val),'\n')
rf_pred = rf.predict(X_val_ohe)
print(confusion_matrix(y_val,rf_pred))
print(classification_report(y_val,rf_pred))
```

```python
# XGBoost
from xgboost import XGBClassifier

xg = XGBClassifier(
    objective="binary:logistic",
    eval_metric="error", 
    n_estimators=10000,  
    random_state=42,
    n_jobs=-1,
    max_depth=5,
    learning_rate=0.1,
    reg_lambda=1,
    min_child_weight =4,
    scale_pos_weight=(y_val == 0).sum() / (y_val == 1).sum()
    )
xg.fit(X_train_ohe, y_train)

print("xg훈련 정확도: ", xg.score(X_train_ohe, y_train))
print("xg검증 정확도: ", xg.score(X_val_ohe, y_val),'\n')
xg_pred = xg.predict(X_val_ohe)
print(confusion_matrix(y_val,xg_pred))
print(classification_report(y_val,xg_pred))
```

```python
# 변수 중요도 plot
def plot_feature_importance(importance, names, topn = 'all'):

    # 변수 중요도와 변수 이름을 정렬해서 df로 저장
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_temp = pd.DataFrame(data)

    fi_temp.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    fi_temp.reset_index(drop=True, inplace = True)

    # 변수가 너무 많으면 보기 어려우니, top N 지정
    if topn == 'all' :
        fi_df = fi_temp.copy()
    else :
        fi_df = fi_temp.iloc[:topn]

    # 그래프 그리기
    plt.figure(figsize=(10,8))
    sns.barplot(x='feature_importance', y='feature_names', data = fi_df)
    plt.xlabel('importance')
    plt.ylabel('feature names')
    plt.grid()

    # 저장한 변수 중요도 df 반환
    return fi_df
```
